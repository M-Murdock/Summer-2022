Converts xml file containing Crosby and Shaeffer vocabulary to a csv file containing only the vocabulary lists

```python
import re
import wget
import pandas as pd
```

### Get the xml file

```python
file_path = 'crosbyshaeffer2.0.xml'
```

Use regex to capture the info after `( *letter* ) Translate:`

```python
capture = False
exercises = []

exercise_num = 0

# get rid of irrelevant lines
with open(file_path) as f:
    for line in f:
        if capture:
            if re.search(r"\([a-z]\)", line):
                capture = False
                exercises[exercise_num] = [i for i in exercises[exercise_num] if i]
                exercise_num += 1
        else:
            if re.search(r"\([a-z]\).*Translate", line):
                capture = True

        # capture everything within the exercise
        if capture:
            # otherwise, grab only the relevant info from within the tags
            match = re.search(r"\>(.*)\<\/", line)
            if match is not None:
                cleaned_match = re.sub(r"<.*>", '', match.group(1))
                cleaned_match = re.sub(r"[&/;]", '', cleaned_match)
                if len(exercises) > exercise_num:
                    exercises[exercise_num].append(cleaned_match)
                else:
                    exercises.append([])
                
print(exercises)
            
```
