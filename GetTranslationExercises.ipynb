Converts xml file containing Crosby and Schaeffer vocabulary to a csv file containing only the vocabulary lists

```python
import re
import wget
import pandas as pd
import csv
```

### Get the xml file

```python
file_path = 'crosbyschaeffer2.0.xml'
```

Use regex to capture the info after `( *letter* ) Translate:`

```python
capture = False
exercises = [[]]

exercise_num = 0

# get rid of irrelevant lines
with open(file_path) as f:
    for line in f:
        if capture:
            # find the end of the translation exercise
            if re.search(r"\([a-z]\)", line):
                capture = False
                exercises[exercise_num] = [i for i in exercises[exercise_num] if i]
                exercise_num += 1
                exercises.append([])
        else:
            # find the beginning of the translation exercise
            if re.search(r"\([a-z]\).*Translate", line):
                capture = True

        # capture everything within the exercise
        if capture:
            # otherwise, grab only the relevant info from within the tags
            match = re.search(r"\>(.*)\<\/", line)
            if match is not None:
                # Remove the 'Translate' line (it's irrelevant for our purposes)
                if not re.search(r"\([a-z]\).*Translate", line):
                    # do more cleaning
                    cleaned_match = re.sub(r"<.*>", '', match.group(1))
                    cleaned_match = re.sub(r"[&/;\"]", '', cleaned_match)
                    # add to the exercises
                    exercises[exercise_num].append(cleaned_match)
  
                
print(exercises)
            
```

```python
ex_no_nums = exercises

for lesson_num in range(0, len(ex_no_nums)):
    for ex_num in range(0, len(ex_no_nums[lesson_num])):
        # remove numbers
        exercises[lesson_num][ex_num] = re.sub(r"[0-9]+", '', ex_no_nums[lesson_num][ex_num])
        # remove periods
        exercises[lesson_num][ex_num] = re.sub(r"[.]+", '', ex_no_nums[lesson_num][ex_num])
        # replace d# with d'
        exercises[lesson_num][ex_num] = re.sub(r"(# )+", '\'', ex_no_nums[lesson_num][ex_num])
        # remove newlines
        exercises[lesson_num][ex_num] = re.sub(r"\n+", '\'', ex_no_nums[lesson_num][ex_num])
        # remove whitespace before and after text
        exercises[lesson_num][ex_num] = exercises[lesson_num][ex_num].strip()
print(ex_no_nums)
```

```python
with open("crosby-schaeffer-translation-exercises.csv","w+") as my_csv:
    csvWriter = csv.writer(my_csv,delimiter=',')
    csvWriter.writerows(exercises)
    
with open("crosby-schaeffer-translation-exercises-no-nums.csv","w+") as my_csv:
    csvWriter = csv.writer(my_csv,delimiter=',')
    csvWriter.writerows(ex_no_nums)
```
